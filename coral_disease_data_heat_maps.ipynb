```python
import pandas as pd
import matplotlib.pyplot as plt
coral_disease_data = pd.read_csv("Guam_corals.csv", index_col=0)
```

```python
pr_coral_data = pd.read_csv("input/NCRMP_Puerto_Rico_corals.csv")
pr_coral_data = pr_coral_data.set_index("species_name")
plot2 = sns.heatmap(pr_coral_data, annot=True, cmap='viridis')
plot2.set(title="Puerto Rico Coral Disease Data Heatmap", ylabel="Coral Species");
```

```python
guam_coral_data = pd.read_csv("input/Guam_corals.csv", index_col=0)
# pr_coral_data = pr_coral_data.set_index("species_name")
plot2 = sns.heatmap(guam_coral_data, annot=True, cmap='viridis')
plot2.set(title="Guam Coral Disease Data Heatmap", ylabel="Coral Species");
```

```python
persian_gulf_coral_data = pd.read_csv("input/Persian_gulf_Hazraty-Kari_et_al.csv", index_col=1)

# Remove weird first column and year
persian_gulf_coral_data = persian_gulf_coral_data.drop(persian_gulf_coral_data.columns[0], axis=1)
persian_gulf_coral_data = persian_gulf_coral_data.drop("year", axis=1)

plot3 = sns.heatmap(persian_gulf_coral_data, annot=True, cmap='viridis')
plot3.set(title="Persian Gulf Coral Disease Data Heatmap", ylabel="Coral Species");
```

```python
isle_data = pd.read_csv("input/Sere_et_al_2015_Isle_de_la_Reunion_Disease_Data.csv", index_col=0)

# Remove region, reef part, and source columns
isle_data = isle_data.drop(["Region", "Reef_Part", "Source"], axis=1)

# Remove None and unknown values from table
isle_data_numeric = isle_data.apply(pd.to_numeric, errors='coerce')

plot3 = sns.heatmap(isle_data_numeric, annot=True, cmap='viridis')
plot3.set(title="Isle de la Reunion Coral Disease Data Heatmap", ylabel="Coral Species");
```

```python
malaysia_data = pd.read_csv("input/Akmal_et_al_Malaysia_data.csv", index_col=0)

# Remove coral family,citation, and doi columns
malaysia_data = malaysia_data.drop(["Coral_family", "citation", "doi"], axis=1)

# Remove None and unknown values from table
#isle_data_numeric = isle_data.apply(pd.to_numeric, errors='coerce')

plot4 = sns.heatmap(malaysia_data, annot=True, cmap='viridis')
plot4.set(title="Malaysia Coral Disease Data Heatmap", ylabel="Coral Species");
```
